```python
import tensorflow as tf

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

dataset_path = "/kaggle/input/datasets/shrutisaxena/yoga-pose-image-classification-dataset/dataset"

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)
```

```python
print(train_ds.class_names[:20])
print("Total Classes:", len(train_ds.class_names))
```

```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras import layers, models

base_model = InceptionV3(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

base_model.trainable = False

model = models.Sequential([
    layers.Rescaling(1./255),
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(107, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)
```

## Objective

The objective of this project is to classify yoga poses using deep transfer learning and provide evidence-based wellness recommendations, potential health-related benefits, and safety precautions for each predicted pose.

## Research Goal

To develop a deep learning-based yoga pose classification system and extend it with an evidence-based wellness recommendation layer.

## Dataset

- Dataset: Yoga Pose Image Classification Dataset
- Total Images: 5,993
- Total Classes: 107

## Methodology

- Image preprocessing
- Transfer Learning using InceptionV3
- Yoga pose classification
- Wellness recommendation generation

## Results

Current validation accuracy after 5 epochs: 27.88%

## Future Work

- Improve classification accuracy
- Add more evidence-based wellness information
- Develop a real-time yoga pose recognition application

## Model Performance

The InceptionV3 transfer learning model was trained on 5,993 yoga pose images belonging to 107 classes.

Training Accuracy (Epoch 5): 21.98%
Validation Accuracy (Epoch 5): 27.88%

The results demonstrate that transfer learning can successfully learn discriminative features for yoga pose classification. Future work will focus on improving accuracy through fine-tuning, data augmentation, and enhanced preprocessing techniques.

```python
for images, labels in val_ds.take(1):

    image = images[0]
    true_label = labels[0]

    prediction = model.predict(tf.expand_dims(image, axis=0))

    predicted_index = np.argmax(prediction)

    predicted_pose = class_names[predicted_index]

    print("Predicted Pose:", predicted_pose)

    plt.figure(figsize=(5,5))
    plt.imshow(image.numpy().astype("uint8"))
    plt.axis("off")
    plt.show()
```

```python
if predicted_pose in benefit_info:
    info = benefit_info[predicted_pose]

    print("\n===== Wellness Recommendation =====")
    print("Pose:", predicted_pose)

    print("\nBenefits:")
    print(info["benefits"])

    print("\nMay Help With:")
    for condition in info["conditions"]:
        print("•", condition)

    print("\nPrecautions:")
    print(info["precautions"])

    print("\nReference:")
    print(info["reference"])
else:
    print("No wellness information available for this pose.")
```
